<span style="font-weight: bold;font-size: 2rem; color: green">Countries winning WC when...
</span>


```python
wc_years = []
for i in wc_df["Winner"]:
    wc_years.append(list(wc_df[wc_df["Winner"]==i]["Year"]))

won_dict = {}
key = 0
for k in wc_df["Winner"]:
    if k not in won_dict.keys():
        won_dict[k] = wc_years[key]
    key = key + 1

# segregating country and their winning values
country_name = list(won_dict.keys())
country_value = list(won_dict.values())
print(country_name)
print(country_value)
```

```python
for j in range(8):
    plt.bar(j, len(country_value[j]),tick_label=country_name[j],hatch="-")
    j = j +1
plt.xticks(range(0,9),country_name)
plt.title("Countries Winning FIFA WorldCup till 2018",color='blue')
plt.xlabel("Countries", color = 'green')
plt.ylabel("Wins",color = 'green')

x=0 
for i in  range(len(country_name)):
    y = 0.72
    for j in range(len(country_value[i])):
        plt.text(x, y, ""+str(country_value[i][j]), color='black', va='center', fontweight='bold',horizontalalignment='center')
        y = y + 0.77
    x = x+1
    
plt.show()
```

<span style="font-weight: bold;font-size: 2.5rem; color: green">Countries with Most WC Goals:
</span>



```python
home_team = matches[['Home Team Name', 'Home Team Goals']].dropna()
away_team = matches[['Away Team Name', 'Away Team Goals']].dropna()
home_team.columns = ['Countries', 'Goals']
away_team.columns = home_team.columns
```

```python
total_goals = home_team._append(away_team, ignore_index = True)
total_goals = total_goals.groupby('Countries',as_index=False).sum()
total_goals=total_goals.sort_values('Goals',ascending=False)[:10]
print(total_goals)
```

```python
ax=sns.barplot(x="Countries",y="Goals",data=total_goals)
```
